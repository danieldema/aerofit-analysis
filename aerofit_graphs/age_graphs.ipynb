```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('age_table.csv')
```

```python
df
```

```python
categories = ['KP281', 'KP481', 'KP781', 'Total']
vals1, vals2, vals3, vals4 = list(df.iloc[0])[1:], list(df.iloc[1])[1:], list(df.iloc[2])[1:], list(df.iloc[3])[1:]
x, w = np.arange(len(categories)), 0.2


fig, ax = plt.subplots(layout = 'constrained', figsize = (10, 5))
ax.bar(x - 3*w/2, vals1, width = w, label = 'KP281')
ax.bar(x - w/2, vals2, width = w, label = 'KP481')
ax.bar(x + w/2, vals3, width = w, label = 'KP781')
ax.bar(x + 3*w/2, vals4, width = w, label = 'Total')

for bars in ax.containers:
    for bar in bars:
        height = bar.get_height()
        if math.isfinite(height):
            ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.0f}',
                    ha='center', va='bottom')

ax.set_xticks(x, labels = ['Age < 30', '30 < Age <= 40', 'Age >= 40', 'Total'])
ax.set_title('Customers of each treadmill type sorted by age')
ax.set_ylabel('Number of customers')
ax.set_xlabel('Age category')
ax.legend()
plt.show()
```

```python
sizes1 = list(df['Total'])[:3]
labels1 = ['KP281', 'KP481', 'KP781']

fig, ax = plt.subplots()
ax.pie(sizes1, labels = labels1, autopct='%1.1f%%')
ax.set_title('Proportion of customers by treadmill')
plt.show()
```

```python
sizes2 = list(df.iloc[3])[1:4]
labels2 = ['Age < 30', '30 <= Age < 40', 'Age >= 40']

fig, ax = plt.subplots()
ax.pie(sizes2, labels = labels2, autopct='%1.1f%%')
ax.set_title('Proportion of customers by age group')
plt.show()
```
